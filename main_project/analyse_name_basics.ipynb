```python
import pandas as pd


df_name_basics = pd.read_csv('../BD_A_IGNORE/name.basics.tsv.gz', compression = 'gzip', sep='\t', on_bad_lines='skip', na_values=['\\N'])

print("My DataFrame shape is : ", df_name_basics.shape, end = '\n\n\n\n')
print('Which columns have missing values ? (Non-null column)')
print(df_name_basics.info())
```
